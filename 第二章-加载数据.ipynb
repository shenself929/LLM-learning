## 加载非结构化数据

- 问题：加载非结构化数据（如文本或图像）。
- 解决：使用Python基础的open函数加载信息：

```python
import requests   进口的请求

# URL to download the txt file from
txt_url = "https://machine-learning-python-cookbook.s3.amazonaws.com/text.txt"

# 获取 txt 文件
r=requests.get(txt_url)

#将其写入本地的 text.txt 文件
with open('text.txt','wb')as f:
    f.write(r.content)
# Read in the file
with open('text.txt','r')as f:
    text=f.read()

print(text)
```

虽然结构化数据可以从CSV、 JSON 或各类数据库中轻松读取，但非结构化数据
则更具挑战性，可能需要后续进行定制化处理。有时使用Python的基本open函数
来打开并读取文件会非常有用——该函数可实现文件的打开及内容读取功能。
